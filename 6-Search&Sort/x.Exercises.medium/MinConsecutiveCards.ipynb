# 2260. Minimum Consecutive Cards to Pick Up

You are given an integer array cards where cards[i] represents the value of the ith card. A pair of cards are matching if the cards have the same value.

Return the minimum number of consecutive cards you have to pick up to have a pair of matching cards among the picked cards. If it is impossible to have matching cards, return -1.

[Leetcode Problem Link](https://leetcode.com/problems/minimum-consecutive-cards-to-pick-up/description/)

### Example 1
Input: cards = [3,4,2,3,2,7]\
Output: 3\
Explanation: We can pick up the cards [2,3,2] which contain a matching pair of cards with value 2. Note that the cards must be picked up in the consecutive order defined in the input, but you can pick the starting point to get the minimum number of cards.

### Example 2
Input: cards = [1,0,5,3]\
Output: -1\
Explanation: There is no way to pick up a set of consecutive cards that contain a pair of matching cards.

## Hash Map Approach

We can use a hash map to track the last index of each card we saw. As we iterate through all the cards, we can use the hash map to quickly compute how many cards we'd need to pick in order to have a pair. For each card, we will also need to update the hash map with the last index we saw the card at.

### Complexity
* Time: O(N); just need to iterate through the cards once
* Space: O(N)
    * worst case: there're no pairs so we end up storing all the cards in the hash map

```python
from typing import List

def minimumCardPickup(cards: List[int]) -> int:
    last_card = {}
    min_distance = -1
    for index in range(len(cards)):
        if cards[index] in last_card:
            if min_distance == -1:
                min_distance = index - last_card[cards[index]] + 1
            else:
                min_distance = min(min_distance, index-last_card[cards[index]]+1)
        last_card[cards[index]] = index
        
    return min_distance

print(minimumCardPickup([95,11,8,65,5,86,30,27,30]))
```
