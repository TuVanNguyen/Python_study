# Number of Provinces

There are n cities. Some of them are connected, while some are not. If city a is connected directly with city b, and city b is connected directly with city c, then city a is connected indirectly with city c.

A province is a group of directly or indirectly connected cities and no other cities outside of the group.

You are given an n x n matrix isConnected where isConnected[i][j] = 1 if the ith city and the jth city are directly connected, and isConnected[i][j] = 0 otherwise.

Return the total number of provinces.

### Example 1
Note that the input is given as an adjacency matrix

Input: isConnected = [[1,1,0],[1,1,0],[0,0,1]]
Output: 2

### Example 2
Input: isConnected = [[1,0,0],[0,1,0],[0,0,1]]
Output: 3


## DFS Recursion Approach
Given an N by N adjacency matrix, we know there are N cities. So we can iterate over each city and perform a depth first search at each city. The depth first search lets us visit all cities directly and indirectly connected with the city, so all those cities are part of the same province.

When we do a depth first search, we want to track each city we've visited, using an array of booleans called `visited`. This ensures that we don't iterate over the connections of a city we've already visited, and it allows us to track whether a city is part of a province we've already visited.

When we iterate over the cities, if a city is not True in `visited`, we know it's part of a new province and increment our count accordingly. Then, we do the depth first search to discover all the cities in that province.

### Analysis
* Time Complexity: $O(N^{2})$
    * Initializing `visited` takes O(N) time
    * The time complexity of `dfs_provinces()` is O(N)
        * It has 1 for loop that iterates over N cities
        * Because of `visited`, the recursion doesn't explore cities more than once. It visits all the cities in O(N) time.
    * `dfs_provinces()` is called inside a for loop that iterates over N, so the entire function is $O(N^{2})$  
* Space Complexity: O(N)
    * the maximum depth of the recursive stack will be O(N)
    * `visited` size is O(N)
    * `visited` exists outside of the dfs function, so it won't be multiplied across the recursion stack

```python
from typing import List


def findCircleNum(isConnected: List[List[int]]) -> int:
    num_cities = len(isConnected)
    visited = [False] * num_cities

    def dfs_provinces(city):
        visited[city] = True
        for connection in range(num_cities):
            if isConnected[city][connection] and not visited[connection]:
                dfs_provinces(connection)

    num_provinces = 0
    for city in range(num_cities):
        if not visited[city]:
            num_provinces += 1
            dfs_provinces(city)

    return num_provinces
```

## DFS Stack Approach

Instead of recursion, you can use a stack to implement depth first search. 

## Analysis
* Time complexity: $O(N^{2})$
    * let's say in the worst case scenario we have a complete graph
    * the outer for loop is O(N)
    * the inner while loop is $O(N^{2})$ on the first iteration, but is O(1) for subsequent iterations
        * first iteration, `dfs_stack` is N length so the while loop iterates over all the cities
        * for the rest of the iterations, `dfs_stack` is empty because we don't add any visited cities to it
        * $T(N) =  N^{2} + N = O(N^{2})$
* Space complexity: O(N)
    * to store `visited`
    * N is max length of dfs_stack, because we don't add visited cities to it

```python
def findCircleNum(isConnected: List[List[int]]) -> int:
    num_cities = len(isConnected)
    visited = [False] * num_cities
    dfs_stack = []
    num_provinces = 0
    
    for city in range(num_cities):
        if not visited[city]:
            num_provinces +=1
            dfs_stack.append(city)
        while dfs_stack:
            new_city = dfs_stack.pop()
            visited[new_city] = True
            for city2 in range(num_cities):
                if isConnected[new_city][city2] and not visited[city2]:
                    dfs_stack.append(city2)
    return num_provinces
```
