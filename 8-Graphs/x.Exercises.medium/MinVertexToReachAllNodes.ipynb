# Minimum Number of Vertices to Reach All Nodes
Given a directed acyclic graph, with n vertices numbered from 0 to n-1, and an array edges where edges[i] = [fromi, toi] represents a directed edge from node fromi to node toi.

Find the smallest set of vertices from which all nodes in the graph are reachable. It's guaranteed that a unique solution exists.

Notice that you can return the vertices in any order.

### Constraints
* 2 <= n <= 10^5
* 1 <= edges.length <= min(10^5, n * (n - 1) / 2)
* edges[i].length == 2
* 0 <= fromi, toi < n
* All pairs (from_i, to_i) are distinct.

## Count Nodes Without Incoming Edges
There are two possibilities for each graph:
- all the nodes are connected such that they form a cycle, so that each node has another pointing to it
- there are nodes that have no other pointing to it, but it points to another node in the graph, So these nodes must be included in the final answer.

To figure this out, we just need to iterate over the edges and track which nodes have others pointing to it.

In the first case, we can return any node. In this case, default to returning [0]. In the second case, you just need to return the list of nodes with none other pointing to them.

### Analysis
* Time Complexity: O(N+E)
    * you need to iterate over all the edges to mark which nodes have others pointing to it
    * you need to iterate over all the nodes to create the final minimum list of nodes
* Space Complexity: O(N)
    * create an array that tracks all the nodes


```python
from typing import List


def findSmallestSetOfVertices(n: int, edges: List[List[int]]) -> List[int]:
    include_vertices = [True]*n
    for edge in edges:
        include_vertices[edge[1]] = False
    
    least_vertices = [i for i in range(n) if include_vertices[i]]
    
    if least_vertices:
        return list(least_vertices)
    else:
        return [0]
```
