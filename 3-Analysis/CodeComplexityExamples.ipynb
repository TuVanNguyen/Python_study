```python
from datetime import date

today = date.today()
year = today.year

if today.year == 2026 and today.day == 5 and today.month == 2:
    print("It's the day!")
else:
    print(f"Today: {today.ctime()}")
```

### Cyclometric complexity score: 2
* base complexity of code is always 1
* we have 1 compound `if` statement 
* 1 + 1 = 2

```python
# From Leetcode Problem 1047: Remove All Adjacent Duplicates In String
def removeDuplicates(s: str) -> str:
    string_stack = []

    for letter in s:
        if len(string_stack) >= 1 and string_stack[-1] == letter :
            string_stack.pop()
        else:
            string_stack.append(letter)
    return "".join(string_stack) 
```

### Cyclometric Complexity score: 3
1. base complexity: 1
1. 1 "for" statement, 
1. 1 compound "if" statement
1. 1 + 1 + 1 = 3
