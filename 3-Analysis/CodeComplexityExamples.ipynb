# Code Complexity Examples

```python
from datetime import date

today = date.today()
year = today.year

if today.year == 2026 and today.day == 5 and today.month == 2:
    print("It's the day!")
else:
    print(f"Today: {today.ctime()}")
```

### Cyclometric complexity score: 4
* base complexity of code is always 1
* we have 1 `if` statement and 2 `and` statements
* 1 + 1 + 2 = 4
