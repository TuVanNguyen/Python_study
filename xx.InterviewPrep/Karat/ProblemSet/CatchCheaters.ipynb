# Catch Cheaters

## Part 1
You are running a classroom and suspect that some of your students are passing around the answers to multiple choice questions disguised as random strings. Your task is to write a function that, given a list of words and a string, finds and returns the word in the list that is scrambled up inside the string, if any exists. There will be at most one matching word. The letters don't need to be in order or next to each other. The letters cannot be reused.

### Example
words = ['cat', 'baby', 'dog', 'bird', 'car', 'ax']\
string1 = 'tcabnihjs'\
find_embedded_word(words, string1) -> cat (the letters do not have to be in order)

string2 = 'tbcanihjs'\
find_embedded_word(words, string2) -> cat (the letters do not have to be together)

string3 = 'baykkjl'\
find_embedded_word(words, string3) -> None / null (the letters cannot be reused)

string4 = 'bbabylkkj'\
find_embedded_word(words, string4) -> baby

string5 = 'ccc'\
find_embedded_word(words, string5) -> None / null

string6 = 'breadmaking'\
find_embedded_word(words, string6) -> bird

### Approach 1: Hashmaps to compare
* Use hashmaps 

#### Complexity
* Time Complexity: $O(W*S)$
    * W = len(words)
    * S = max length of each string in words
    * this is assuming processing `words` is much more than scanning `code` once
    * if `code` is sufficiently long, time complexity could also be written as O(W*S + C) where C = len(code)
* Space: O(W)
    * W accounts for each `word` in `word_maps`
    * because the hashmap for each word will be at most 26, it is a constant value and doesn't need to be factored in
    * length of `code` wouldn't factor in because `string_char_map` length is fixed at 26, so it's a constant

```python

def solve(words, code):
    word_map = {word: {} for word in words}
    for word in words:
        for c in word:
            if c not in word_map[word]:
                word_map[word][c] = 1
            else:
                word_map[word][c] +=1

    string_char_map = {}
    for c in code:
        if c not in string_char_map:
            string_char_map[c] = 1
        else:
            string_char_map[c] += 1
    
    for word in words:
        found = True
        for c in word:
            if not string_char_map.get(c) or word_map[word][c] > string_char_map[c]:
                found = False
        if found == True:
            return word

    return None

words = ['cat', 'baby', 'dog', 'bird', 'car', 'ax']
string1 = 'tcabnihjs'
string2 = 'tbcanihjs'
string3 = 'baykkjl'
string4 = 'bbabylkkj'
string5 = 'ccc'
string6 = 'breadmaking'

print(solve(words, string1))
print(solve(words, string2))
print(solve(words, string3))
print(solve(words, string4))
print(solve(words, string5))
print(solve(words, string6))
```

## Part 2
After catching your classroom students cheating before, you realize your students are getting craftier and hiding words in 2D grids of letters. The word may start anywhere in the grid, and consecutive letters can be either immediately below or immediately to the right of the previous letter.

Given a grid and a word, write a function that returns the location of the word in the grid as a list of coordinates. If there are multiple matches, return any one.

### Approach 1: Recursion
I'm going to start by traversing the grid until I find a character that starts with the same letter as my word. From there, I'll need to do a recursive step that searches for the rest of the word by looking at the next column or the next row.

Use a helper function to do the recursive step, so that there is separation from the loop that is traversing the entire grid. Otherwise, I can easily get disjointed points in my final answer.

#### Analysis
* Time Complexity: $O(R*C*2^{W})$
    * R = # rows, C = # columns, W = word length
    * traversing the entire grid is T(R*C)
    * Doing the recursive search is $T(2^W)$, and I do it each time I find a matching character as I traverse the grid

```python
def recursive_search(grid, word, start_row = 0, start_col = 0):
    if not word:
        return []
    if start_row >= len(grid) or start_col >= len(grid[start_row]):
        return []
    if word[0] == grid[start_row][start_col]:
        answer = [(start_row, start_col)]
        next_row = recursive_search(grid, word[1:], start_row=start_row+1, start_col=start_col)
        next_col = recursive_search(grid, word[1:], start_row=start_row, start_col=start_col+1)
        if next_row:
            answer+=next_row
        elif next_col:
            answer += next_col
        if len(answer) == len(word):
            return answer
    return []
    

def find_word_location(grid, word, start_row = 0, start_col = 0):
    if not word:
        return []
    for row in range(start_row,len(grid)):
        for col in range(start_col, len(grid[row])):
            if grid[row][col] != word[0]:
                continue
            answer = recursive_search(grid, word, start_row=row, start_col=col)
            if len(answer) == len(word):
                return answer

    return []
grid1 = [
        ['c', 'c', 'x', 't', 'i', 'b'],
        ['c', 'c', 'a', 't', 'n', 'i'],
        ['a', 'c', 'n', 'n', 't', 't'],
        ['t', 'c', 's', 'i', 'p', 't'],
        ['a', 'o', 'o', 'o', 'a', 'a'],
        ['o', 'a', 'a', 'a', 'o', 'o'],
        ['k', 'a', 'i', 'c', 'k', 'i'],
    ]
word1 = "catnip"
word2 = "cccc"
word3 = "s"
word4 = "bit"
word5 = "aoi"
word6 = "ki"
word7 = "aaa"
word8 = "ooo"
grid2 = [['a']]
word9 = "a"


print(find_word_location(grid1, word1)) # [ (1, 1), (1, 2), (1, 3), (2, 3), (3, 3), (3, 4) ]
print(find_word_location(grid1, word2)) # [(0, 1), (1, 1), (2, 1), (3, 1)]
print(find_word_location(grid1, word3)) # [(3, 2)]
print(find_word_location(grid1, word4)) # [(0, 5), (1, 5), (2, 5)]
```
