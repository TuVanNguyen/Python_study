# Prefix Sum
* typically used on arrays of numbers
* create an array `prefix` such that `prefix[i]` is the sum of all elements up to index i

### Why Use Prefix Sum?
* Be able to query sum of all subarrays in O(1) time , at the expense of n memory
* Prefix array typically takes only O(n) time to create
* can improve an algorithm's time complexity by a factor of O(n)

### Important properties
* in prefix, sum of subarrays from i to j: $ prefix[j] - prefix[i] + nums[i] $

```python
# Implementation of prefix sum

from typing import List

def prefix_sum(nums: List[int]):
    prefix = [nums[0]]
    for num in nums[1:]:
        prefix.append(prefix[-1] + num)
    return prefix

print(prefix_sum([3,6,2,8,1,4,1,5]))
```
