# Online Stock Span

Design an algorithm that collects daily price quotes for some stock and returns the span of that stock's price for the current day.

The span of the stock's price in one day is the maximum number of consecutive days (starting from that day and going backward) for which the stock price was less than or equal to the price of that day.

For example, if the prices of the stock in the last four days is [7,2,1,2] and the price of the stock today is 2, then the span of today is 4 because starting from today, the price of the stock was less than or equal 2 for 4 consecutive days.
Also, if the prices of the stock in the last four days is [7,34,1,2] and the price of the stock today is 8, then the span of today is 3 because starting from today, the price of the stock was less than or equal 8 for 3 consecutive days.

[Leetcode Link](https://leetcode.com/problems/online-stock-span/description/)

### Example 1
Input:\
["StockSpanner", "next", "next", "next", "next", "next", "next", "next"]\
[[], [100], [80], [60], [70], [60], [75], [85]]\

Output:[null, 1, 1, 1, 2, 1, 4, 6]

### Constraints:
* 1 <= price <= 105
* At most 104 calls will be made to next.

## Approach 1: Brute Force
* I'm going to store all the prices as I see them in a list, on my StockSpanner object
* Iterating backwards on the list, I count the number of consecutive prices that are lower than the current one
* To optimize a bit, I'll also store the previous spans. So for each previous lower price I encounter, I'll add its span with my current one and jump further back in my list of prices by the previous price's span.

### Analysis
* Time Complexity: O(1) per next() call
    * O(N) for a total of N calls
    * Because I'm only visiting each span once over the course of N calls, it amortizes to O(1) for each next() call
* Space Complexity: O(N)
    * for the two tracking lists I have
* Code Complexity: 
    * Cyclometric complexity: 2 (very good!)
        * 1 base complexity
        * 1 while loop
        * 1 `and` operator 


```python
class StockSpanner(object):

    def __init__(self):
        self.track_prices = []
        self.track_spans = []
        

    def next(self, price):
        """
        :type price: int
        :rtype: int
        """
        span = 1
        check = len(self.track_prices) - 1
        while check >= 0 and self.track_prices[check] <= price:
            span += self.track_spans[check]
            check -= (self.track_spans[check])
        self.track_prices.append(price)
        self.track_spans.append(span)
        return span
```

## Approach 2: Monotonic Stack
To make it a bit more readable, we can also use a monotonic stack instead of doing the jump-back. The stack stores prices that are greater than the previous price whose span we calculated last time. This way, we keep track of prices that aren't included in the previous span, but may be included when calculating the current span.

### Analysis
* Time Complexity: O(1) per next() call
    * O(N) for a total of N calls
    * Because I'm only checking each previous span once over the course of N calls, it amortizes to O(1) for each next() call
* Space Complexity: O(N)
    * for the two tracking lists I have
* Code Complexity: 
    * Cyclometric complexity: 2 (very good!)
        * 1 base complexity
        * 1 while loop
        * 1 `and` operator 

```python
class StockSpanner(object):

    def __init__(self):
        self.price_stack = []
        #self.track_spans = []
        

    def next(self, price):
        """
        :type price: int
        :rtype: int
        """
        span = 1
        while self.price_stack and self.price_stack[-1][0] <= price:
            span += self.price_stack[-1][1]
            self.price_stack.pop()
        self.price_stack.append((price,span))
        return span
```
