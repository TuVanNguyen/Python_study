# Daily Temperatures

Given an array of integers temperatures represents the daily temperatures, return an array answer such that answer[i] is the number of days you have to wait after the ith day to get a warmer temperature. If there is no future day for which this is possible, keep answer[i] == 0 instead.

[Leetcode Link](https://leetcode.com/problems/daily-temperatures/description/)

### Example 1
Input: temperatures = [73,74,75,71,69,72,76,73]\
Output: [1,1,4,2,1,1,0,0]

### Example 2
Input: temperatures = [30,40,50,60]\
Output: [1,1,1,0]

### Example 3
Input: temperatures = [30,60,90]\
Output: [1,1,0]

### Constraints
* $1 <= temperatures.length <= 10^{5}$
* 30 <= temperatures[i] <= 100

## Brute Force Approach
For each temperature, we can simply compare it to the rest of the list. We'll keep looking for a day with a higher temperature, stop when or if we find it, record the day, then move on to the next temperature.

### Complexity Analysis
* Time Complexity: $O(N^{2})$
    * In the worst case scenario, the higher temperature is always the last day
    * So the number of comparisons will be a sum of natural numbers: T(n) = $\sum_{i=0}^{n}i = \frac{n(n+1)}{2} = \frac{1}{2}n^2 + \frac{1}{2}n$ 
    * The time complexity of a sum of natural numbers is $O(N^{2})$
    * we can also quickly tell the time complexity will be $O(N^{2})$ because we'll need double nested for loops  
* Space Complexity: O(N)
    * we'll just need another array of N length to store the answer. That's the best it can be.
* Code Complexity: Maccabe score 4
    * 1 base
    * 2 for loops
    * 1 if conditional
    * 1 + 2 + 1 = 4
    * Overall very simple to follow and understand




```python
from typing import List


def dailyTemperatures(temperatures: List[int]) -> List[int]:
    answer = [0] * len(temperatures)
    for temp_index in range(len(temperatures)):
        for next_index in range(temp_index + 1, len(temperatures)):
            if temperatures[temp_index] < temperatures[next_index]:
                answer[temp_index] = next_index - temp_index
                break
    return answer
```

# Monotonic Stack Approach
Alternatively, we can use a monotonic stack to track the days that we haven't figured out the next warmest day for them yet. We also want to store the index of the day instead of the temperature value itself, because it would make it much easier to calculate how far apart the days are when we encounter a warmer day.

The monotonic stack will contain the indexes of days in order of decreasing temperature. So, the top of the stack is always the day with the lowest temperature. If the next temperature is higher, we keep popping the colder days from the stack and record how far apart the two days were. Then, we append the next day. This way, we make sure the order of the stack is maintained.

### Complexity
* Time Complexity: O(N)
    * We only have to do one pass through the array
* Space Complexity: O(N)
    * The stack can be up to N in length, and we also need the answer array
* Code Complexity:  McCabe score 3
    * 1 base
    * 1 for loop
    * 1 while loop
    * 1 + 1 + 1 = 3
    * overall, the flow is simple and easy to follow. It's ideal for interviews and on-the-job. 


```python
from typing import List


def dailyTemperatures(temperatures: List[int]) -> List[int]:
    answer = [0] * len(temperatures)
    tracker  = []
    for temp_index in range(len(temperatures)):
        while tracker and temperatures[temp_index] > temperatures[tracker[-1]]:
            last_index = tracker.pop()
            answer[last_index] = temp_index - last_index
        tracker.append(temp_index)
    return answer
```
