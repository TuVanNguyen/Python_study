## 4.2 Linear Structures
* A collection where once an item is added, the item stays in that position relative to other elements that came before or after it
* includes stacks, queues, deques, and lists 
* has two ends where (depending on the type of linear structure), an item can be added or removed

### Why know these linear structures?
* some of the most useful data structures in computer science because they're used in many algorithms

## 4.3 The Stack
* adding and subtracting items always takes place at the same end known as the "top"
* the other end is "the base"
* Ordering principle: LIFO (last-in first-out)
* Important property: removal is reverse order of insertion
    * One use is using the "back" button on web browsers to go to the last page visited: the button retrieves pages in reverse order to when you visited them
* See [Stack.py](Stack.py) for full implementation of a stack

### Stack implemented with dynamic array (Python list)
| Operation | Time Complexity | Implementation |
|-----------|-----------------|----------------|
| create | O(1) | stack = [] |
| push | O(1) | stack.append(1) |
| pop | O(1) | stack.pop() |
| get at index | O(1) | stack[0] |
| search | O(n) | stack.index(1) |

## 4.10 Queue
* an collection where new items are added to the "rear" end, and items are removed at the "front" end
* Ordering Principle: FIFO (first-in first-out)

### Queue implemented as dynamic array (Python list)

|Operation | Time Complexity | Implementation |
|----------|-----------------|----------------|
| dequeue: remove front element | O(N) | queue = queue[1:] |
| enqueue: add element to back | O(1) | queue.append("x") |

* Note if you want better performance with a queue, you'll need to implement it as a doubly linked list or use `collections.dequeue`


## 4.10.1 Monotonic Stacks and Queues
* definition: describes a function or quantity varying in such a way that it either never decreases or never increases
    * i.e it's always sorted, whether in ascending or descending order

### How it's implemented in stacks and queues
* when adding a new element, you would need to remove elements and add them back in to maintain the order

| Method | Time Complexity | 
|--------|-----------------|
| add element | O(N) |

### Uses
* you want to easily track maximum or minimum element as the window in the stack or queue changes
* for problems where you want to find the "next" element in some order

## 4.15 Deque
* double-ended queue, meaning you can add or remove items from either end
* Ordering Principle: user's discretion

```python
import collections

queue = collections.deque()
queue = collections.deque([1, 2, 3])

queue.append(4)
queue.append(5)
queue.popleft() # remove 1
queue.popleft() # remove 2
queue.pop() # remove 5
print(queue)
```

## 4.21 Linked List
* A bunch of nodes that has a pointer to the next item

### Why used linked list?
* in other languages, it has the advantage of dynamic sizing
* however in python, we already have that benefit with standard list 
* it's just useful to understand linked list algorithms for coding interviews

### Types of Linked List
* Singly Linked List: each node has pointer to next node
* Doubly Linked List: each node has a pointer to the next node, and another pointer to the previous node


### Operations

| Operation | Time Complexity | explaination |
|-----------|-----------------|--------------|
| get $n^{th}$ element | O(n) | you need to iterate to the $n^{th}$ item in the linked list to access that element | 
| add element at $n^{th} position without a reference| O(n) | you need to iterate to the $n^{th}$ item in the linked list to access that position |
| remove element at $n^{th} position without a reference| O(n) | you need to iterate to the $n^{th}$ item in the linked list to access that position|
