# Sliding Window Maximum
You are given an array of integers nums, there is a sliding window of size k which is moving from the very left of the array to the very right. You can only see the k numbers in the window. Each time the sliding window moves right by one position.

Return the max sliding window.

[Leetcode Link](https://leetcode.com/problems/sliding-window-maximum/description/)

### Example 1
Input: nums = [1,3,-1,-3,5,3,6,7], k = 3\
Output: [3,3,5,5,6,7]

### Example 2
Input: nums = [1], k = 1\
Output: [1]

### Constraints
* $1 <= nums.length <= 10^{5}$
* $-10^{4} <= nums[i] <= 10^{4}$
* 1 <= k <= nums.length

## Brute Force Approach

You can simply look for the largest element in each window by comparing all the numbers in every window.

### Complexity
* Time Complexity: O(N*K)
    * We iterate through the entire list of N length once
    * for every possible window of length K, we iterate over that window to find the maximum element
    * worst case scenario: K = (N/2), which gives us $\lceil\frac{N}{2}\rceil + 1$ windows
        * T(N) = $\frac{N}{2} * (\frac{N}{2} + 1)$ = $\frac{N^{2}}{4} + \frac{N}{2}$ = $O(N^{2})$
