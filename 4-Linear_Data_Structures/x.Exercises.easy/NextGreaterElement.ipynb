# Next Greater Element
The next greater element of some element x in an array is the first greater element that is to the right of x in the same array.

You are given two distinct 0-indexed integer arrays nums1 and nums2, where nums1 is a subset of nums2.

For each 0 <= i < nums1.length, find the index j such that nums1[i] == nums2[j] and determine the next greater element of nums2[j] in nums2. If there is no next greater element, then the answer for this query is -1.

Return an array ans of length nums1.length such that ans[i] is the next greater element as described above.

### Example 1
Inputs: nums1 = [4,1,2], nums2 = [1,3,4,2]\
Output: [-1,3,-1]

### Example 2
Input: nums1 = [2,4], nums2 = [1,2,3,4]
Output: [3,-1]


## Approach 1: Brute Force

### Analysis
* Time Complexity: O(M*N)
    * M = len(nums1), N = len(nums2)
* Space Complexity: O(M)
    * M = len(nums1)


```python
from typing import List


def nextGreaterElement(nums1: List[int], nums2: List[int]) -> List[int]:
    answer = [-1] * len(nums1)
    for i in range(len(nums1)):
        found = False
        for j in range(len(nums2)):
            if nums2[j] == nums1[i]:
                found = True
            elif found and nums2[j] > nums1[i]:
                answer[i] = nums2[j]
                break
    return answer
        
print(nextGreaterElement([4,1,2], [1,3,4,2]))
```

# Approach 2: Use Stack and Hashmap
Use the hashmap to track the next greater element for each element as you encounter them as you traverse across `nums2`.

Use the stack to track the elements for which you still need to find the next greater element.

In the end, you can traverse across `num1` and do a lookup for each element in the hashmap to get your answer.

### Analysis
* Time Complexity: O(N)
    * breaking down the operations, it would be T(M+N)
    * N > M, so it simplifies to O(2N) = O(N)
* Space Complexity: O(N)
    * the hashmap will have N entries, and the stack will at most be N length 

```python
def nextGreaterElement(nums1: List[int], nums2: List[int]) -> List[int]:
    greater_elements = {}
    elements = []
    for i in range(len(nums2)):
        while elements and nums2[i] > elements[-1]:
            last = elements.pop()
            greater_elements[last] = nums2[i]
        elements.append(nums2[i])

    while elements:
        greater_elements[elements.pop()] = -1

    answer = [greater_elements[num] for num in nums1]
    return answer
```
